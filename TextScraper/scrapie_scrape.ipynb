```python
collection = open('chernobyl.txt', 'r')
articles = collection.read()
collection.close()

# seperate articles into list
lines = re.sub('<doc>', ' ', lines)
lines = articles.split('</doc>')[:-1]
lines[1]
```

```python
soup = BeautifulSoup(articles,'lxml')
```

```python
alltext = soup.text
```

```python
all_articles = soup.findAll('text')
first_article = all_articles[0]
second_article = str(all_articles[1]).split()
third_article = all_articles[2]
len(second_article)
type(all_articles)
```

```python
class Article:
    def __init__(self, dictionary):
        for i, j in dictionary.items():
            setattr(self, i, j)
```

```python

for i in range(len(lines)):
	# Seperate article items based on tags
	lines[i] = re.sub('<.?p>|</.*?>', ' ', lines[i])
	lines[i] = re.sub('\n', '', lines[i])
	lines[i] = lines[i][2:]
	lines[i] = lines[i].split(' <')
    
    # Create Article object of article items
	tempDict = {}
	for j in range(0, len(lines[i]), 2):
		tempDict[lines[i][j]] = lines[i][j+1]
	lines[i] = Article(tempDict)
```
